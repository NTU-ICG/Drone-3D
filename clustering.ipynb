# 3D and 2D Plot visualisation


```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D 
import plotly.graph_objs as go 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objs as go
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
 
# Assuming lidar_positions is your numpy array of shape [n, 3] 
# Replace it with your actual data 
lidar_positions = np.load("1706255123.300333.npy")  # Example random lidar positions 
 
# Matplotlib 3D plot 
fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d') 
ax.scatter(lidar_positions[:, 0], lidar_positions[:, 1], lidar_positions[:, 2], s=1) 
ax.set_xlabel('X') 
ax.set_ylabel('Y') 
ax.set_zlabel('Z') 
ax.set_title('Matplotlib 3D Scatter Plot') 
plt.show() 
 
# Matplotlib 2D plot 
fig = plt.figure() 
ax = fig.add_subplot() 
ax.scatter(lidar_positions[:, 0], lidar_positions[:, 1], s=1) 
ax.set_xlabel('X') 
ax.set_ylabel('Y') 
ax.set_title('Matplotlib 3D Scatter Plot') 
plt.show() 
 
# Matplotlib 2D plot 
fig = plt.figure() 
ax = fig.add_subplot() 
ax.scatter(lidar_positions[:, 0], lidar_positions[:, 1], s=1) 
ax.set_xlabel('X') 
ax.set_ylabel('Y') 
ax.set_xlim((-5, 5)) 
ax.set_ylim((-5, 5)) 
ax.set_title('Matplotlib 3D Scatter Plot') 
plt.show() 
 
# Plotly 3D plot 
trace = go.Scatter3d( 
    x=lidar_positions[:, 0], 
    y=lidar_positions[:, 1], 
    z=lidar_positions[:, 2], 
    mode='markers', 
    marker=dict( 
        size=1, 
        color='rgb(0, 0, 255)', 
        opacity=0.8 
    ) 
) 
 
layout = go.Layout( 
    title='Plotly 3D Scatter Plot', 
    scene=dict( 
        xaxis=dict(title='X'), 
        yaxis=dict(title='Y'), 
        zaxis=dict(title='Z') 
    ) 
) 
 
fig = go.Figure(data=[trace], layout=layout) 
fig.show()
```

# Testing out - K-Means with 2,3,4,5 centers

```python
n_clusters_options = [2, 3, 4, 5]  # Varying the number of clusters
clusterings = {}

for n_clusters in n_clusters_options:
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(lidar_positions)
    clusterings[n_clusters] = kmeans

# Select a clustering to visualize (e.g., with 3 clusters)
selected_clustering = clusterings[3]

# Matplotlib 3D plot with clusters
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(lidar_positions[:, 0], lidar_positions[:, 1], lidar_positions[:, 2], c=selected_clustering.labels_, s=1)
ax.scatter(selected_clustering.cluster_centers_[:, 0], selected_clustering.cluster_centers_[:, 1], selected_clustering.cluster_centers_[:, 2], s=100, c='red')  # Red dots for cluster centers
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Matplotlib 3D Scatter Plot with Clusters')
plt.colorbar(scatter)
plt.show()

# Plotly 3D plot with clusters
trace = go.Scatter3d(
    x=lidar_positions[:, 0],
    y=lidar_positions[:, 1],
    z=lidar_positions[:, 2],
    mode='markers',
    marker=dict(
        size=1,
        color=selected_clustering.labels_,
        colorscale='Viridis',  # Color scale can be adjusted
        opacity=0.8
    )
)

centers_trace = go.Scatter3d(
    x=selected_clustering.cluster_centers_[:, 0],
    y=selected_clustering.cluster_centers_[:, 1],
    z=selected_clustering.cluster_centers_[:, 2],
    mode='markers',
    marker=dict(
        size=5,
        color='rgb(255, 0, 0)',  # Red color for cluster centers
        opacity=1
    )
)

layout = go.Layout(
    title='Plotly 3D Scatter Plot with Clusters',
    scene=dict(
        xaxis=dict(title='X'),
        yaxis=dict(title='Y'),
        zaxis=dict(title='Z')
    )
)

fig = go.Figure(data=[trace, centers_trace], layout=layout)
fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.spatial.distance import cdist
# Load your lidar data
# Replace this with the actual data you want to cluster
lidar_positions = np.load("1706255123.300333.npy")

# Assuming lidar_positions is a [n, 2] numpy array where n is the number of points
# and each point is represented as (x, y).

# Determine the optimal number of clusters using Silhouette Score
range_n_clusters = list(range(2, 6))
best_score = -1
best_n_clusters = 0
best_kmeans = None

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=10).fit(lidar_positions)
    cluster_labels = kmeans.labels_
    silhouette_avg = silhouette_score(lidar_positions, cluster_labels)
    
    if silhouette_avg > best_score:
        best_score = silhouette_avg
        best_n_clusters = n_clusters
        best_kmeans = kmeans

print(f"Best number of clusters: {best_n_clusters}")

# KMeans with the optimal number of clusters found
kmeans = best_kmeans
cluster_centers = kmeans.cluster_centers_

# Plotting the results
plt.figure(figsize=(10, 8))
plt.scatter(lidar_positions[:, 0], lidar_positions[:, 1], c=kmeans.labels_, cmap='viridis', s=1)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', s=50, alpha=0.75) # cluster centers
plt.title("KMeans Clustering")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

```

# Testing out: DBSCAN Prediction (Poor)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import DBSCAN

# Load your lidar data
# Replace this with the actual data you want to cluster
lidar_positions = np.load("1706255123.300333.npy")

# DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=10)  # eps and min_samples need to be tuned for your dataset
dbscan.fit(lidar_positions)

# The labels_ attribute will be -1 for outliers
outlier_mask = dbscan.labels_ == -1
core_mask = ~outlier_mask

# Core samples will be those that are not outliers
core_samples = lidar_positions[core_mask]
outliers = lidar_positions[outlier_mask]

# 3D scatter plot using Matplotlib
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
# Plot core points
ax.scatter(core_samples[:, 0], core_samples[:, 1], core_samples[:, 2], c='blue', s=1)
# Plot outliers which could be the drone location
ax.scatter(outliers[:, 0], outliers[:, 1], outliers[:, 2], c='red', s=50, alpha=0.75)
ax.set_title("DBSCAN Clustering")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

# K Means (predicting the coordinates with expected coordinates known)

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


# Load your lidar data
# Replace this with the actual data you want to cluster
lidar_positions = np.load("1706255123.300333.npy")

# Determine the optimal number of clusters using the silhouette score
silhouette_scores = []
for n_cluster in range(2, 5):
    kmeans = KMeans(n_clusters=n_cluster).fit(lidar_positions)
    score = silhouette_score(lidar_positions, kmeans.labels_)
    silhouette_scores.append(score)
    print(f"Number of clusters: {n_cluster}, Silhouette Score: {score}")

best_n_clusters = silhouette_scores.index(max(silhouette_scores)) + 2  # +2 because range starts at 2
print(f"Best number of clusters: {best_n_clusters}")

# Apply KMeans with the determined number of clusters
kmeans = KMeans(n_clusters=best_n_clusters)
kmeans.fit(lidar_positions)

# Assign each point to the nearest cluster centroid
clusters = kmeans.predict(lidar_positions)

# Now visualize the clusters
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(lidar_positions[:, 0], lidar_positions[:, 1], lidar_positions[:, 2], c=clusters, s=1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('KMeans Clustering')
plt.legend(*scatter.legend_elements(), title="Clusters")
plt.show()

# The centroid of the cluster that is potentially the drone can be found in kmeans.cluster_centers_
# If you have an expected altitude range for the drone, you can filter out centroids that are not within that range.
possible_drone_positions = kmeans.cluster_centers_

# Optionally, if you know the expected coordinates, you can find the closest cluster to that position
# Replace 'expected_position' with the actual expected x, y, z coordinates
```

```python
expected_position = np.array([12.69621039,-19.60909913,43.36182919])  # example expected position
distances = np.linalg.norm(possible_drone_positions - expected_position, axis=1)
drone_cluster_index = np.argmin(distances)
drone_position = possible_drone_positions[drone_cluster_index]
print(f"Predicted position of the drone: {drone_position}")
```
